```python
movies['genres']  # observe that one movie has multiple genres..so we will go to each movie genres and splits them
```

```python
# First we need to convert them into a str
movies['genres'].str
```

```python
# Then split
movies['genres'].str.split('|')
```

```python
# Now check if Action is present
movies['genres'].str.split('|').apply(lambda x:'Action' in x)
```

```python
mask1 = movies['genres'].str.split('|').apply(lambda x:'Action' in x)
mask2 = movies['imdb_rating'] > 7.5
# Combine both
movies[mask1 & mask2]
```

```python
movies[mask1 & mask2].shape[0]
```

```python
# Alternate way
mask11 = movies['genres'].str.contains('Action')
mask22 = movies['imdb_rating'] > 7.5

movies[mask11 & mask22].shape[0]
```

# Group Attributes and Methods

```python
# Find toal number of group --->
len(movies.groupby('Genre'))
```

```python
movies.groupby('Genre').size()
```

```python
# first()/second()....nth()
movies.groupby('Genre').first()
```

```python
movies.groupby('Genre').last()
```

```python
movies.groupby('Genre').nth(6).set_index('Genre')
```

```python
# get_group  ---> filtering
movies.groupby('Genre').get_group('Fantasy').set_index('Genre')
```

```python
movies[movies['Genre'] == 'Fantasy'].set_index('Genre')
```

```python
movies.groupby('Genre').get_group('Horror').set_index('Genre')
```

```python
movies[movies['Genre'] == 'Horror'].set_index('Genre')
```

```python
# describr
movies.groupby('Genre').describe()
```

```python
# sample
movies.groupby('Genre').sample(2,replace=True).set_index('Genre')
```

```python
# agg method
movies.groupby('Genre')[movies.select_dtypes(include='number').columns].sum()
```

```python
# using agg we can apply different aggregrate fucntions on different columns
movies.groupby('Genre')[movies.select_dtypes(include='number').columns].agg(
{
    'Runtime':'mean',
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# using agg we can apply different aggregrate fucntions on different columns
movies.groupby('Genre')[movies.select_dtypes(include='number').columns].agg(
{
    'Runtime':['mean','sum'],
    'IMDB_Rating':'mean',
    'No_of_Votes':'sum',
    'Gross':'sum',
    'Metascore':'min'
})
```

```python
# Using list
movies.groupby('Genre')[movies.select_dtypes(include='number').columns].agg(['min','max','sum','mean'])
```
